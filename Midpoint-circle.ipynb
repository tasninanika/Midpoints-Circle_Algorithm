```python
import matplotlib.pyplot as plt


def midpoint_circle(cx, cy, r):
    x, y = 0, r
    d = 1 - r
    points = []
    
    def plot_circle_points(cx, cy, x, y):
        points.extend([(cx + x, cy + y), (cx - x, cy + y), (cx + x, cy - y), (cx - x, cy - y),
                       (cx + y, cy + x), (cx - y, cy + x), (cx + y, cy - x), (cx - y, cy - x)])
    
    plot_circle_points(cx, cy, x, y)
    
    while x < y:
        x += 1
        if d < 0:
            d += 2 * x + 1
        else:
            y -= 1
            d += 2 * (x - y) + 1
        plot_circle_points(cx, cy, x, y)
    
    return points

def plot_graph(circle_points):
    
    if circle_points:
        x_vals, y_vals = zip(*circle_points)
        plt.scatter(x_vals, y_vals, color='r', s=40)
    
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("Midpoint Circle Algorithm")
    plt.grid(True)
    plt.axis("equal")
    plt.show()

# Example usage
x1, y1 = 2, 3
x2, y2 = 10, 7
cx, cy, r = 5, 5, 4
circle_points = midpoint_circle(cx, cy, r)
plot_graph(circle_points)
```
